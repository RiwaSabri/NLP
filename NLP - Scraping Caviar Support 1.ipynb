```python
import pickle
import os
import pandas as pd
# Setting variables to be used in format string command below
#tweet_count = 500
text_query = "@CaviarSupport"
since_date = "2010-01-01"
until_date = "2022-10-18"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --since {} twitter-search "{} until:{}"> caviar.json'
          .format(since_date, text_query, until_date))
```

```python
df = pd.read_json('caviar.json', lines = True)
```

```python
df
```

```python
def return_username(user):
    return user['username']
```

```python
df['user']=df.user.apply(lambda x:return_username(x))
```

```python
df=df[(df['user']!='CaviarSupport')& (df['inReplyToTweetId'].isna())]
```

```python
df
```

```python
with open('caviar.pickle', 'wb') as handle:
    pickle.dump(df, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
